DataFrame in Python Pandas

```python
import pandas as pd
l=[1,2,3,4,5,6,7]
var=pd.DataFrame(l)
print(type(var))
```

```python
d={"a":[1,2,3,4],"b":[5,6,7,8],"c":[9,10,11,12]}
var1=pd.DataFrame(d,columns=["b","a","c"],index=["r1","r2","r3","r4"])
print(type(var1))
print(var1)
print(var1["a"][3])
```

```python
list_1=[[1,2,3,4,5,6],[11,12,13,14,15,16],[21,22,23,24,25,26]]
var2=pd.DataFrame(list_1,columns=["a","b","c","d","e","f"],index=["r1","r2","r3"])
print(type(var2))
print(var2)
```

```python
sr={"s":pd.Series([1,2,3,4],index=["a","b","c","d"]),
    "t":pd.Series([5,6,7,8],index=["a","b","c","d"])}
var3=pd.DataFrame(sr)
print(type(var3))
print(var3)
print(sr)
```
