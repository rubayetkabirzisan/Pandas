Pivot Table and Melt Function 

```python
import pandas as pd
var=pd.DataFrame({'A':['foo','foo','foo','bar','bar','bar'],
                  'B':['one','one','two','two','one','one'],'C':['x','y','x','y','x','y'],})

pd.melt(var,id_vars=['A'],var_name="python",value_name="language")
```

```python
var1=pd.DataFrame({'A':['foo','foo','foo','bar','bar','bar'],
                  'B':['one','one','two','two','one','one'],'C':['x','y','x','y','x','y'],'D':[1,2,3,4,5,6],'E':[7,8,9,10,11,12]})
# var1.pivot(index=['D'],columns=['A'],values='E')
pivot_table=var1.pivot_table(index=['D'],values=['A','B'],aggfunc='sum',margins=True)
print(pivot_table)

```
