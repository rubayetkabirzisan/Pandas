dropna and fillna

```python
import pandas as pd
var = pd.read_csv("/e:/Pandas/Video 9/demo.csv")
var
```

```python
var.dropna()
```

```python
var.dropna(axis=1)
var.dropna(how="any")
var.dropna(how="all")
var.dropna(subsert=["CHANGE"]) 
var.dropna(inplace=True)
var.dropna(thresh=1)


```

```python
var.fillna("No Value")
var.fillna({"SYMBOL":"python","SECURITY":"C","TODAY - VOLUME":"JAVA"})
var.fillna(method="ffill")
var.fillna(method="ffill",axis=1)
var.fillna(method="bfill")
var.fillna(12,inplace=True)
var.fillna("python",limit=2)
```
