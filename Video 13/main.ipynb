Join and Append DataFrames

```python
import pandas as pd

var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]},index=["a","b","c"])
var2 = pd.DataFrame({'C': [7, 8], 'B': [10, 11]},index=["a","b"])

var1.join(var2,how="outer",lsuffix="_right",rsuffix="_left")
```

```python
# Append DataFrames
appended_df = pd.concat([var1, var2], ignore_index=True)
print("Appended DataFrame:")
print(appended_df)
```
