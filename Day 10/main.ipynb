Replace and interpolate

```python
import pandas as pd
var = pd.read_csv("E:\Pandas\Video 10\demo.csv")
print(var)
```

```python
var.replace(to_replace ="HINDCON",value = "BHARATCON")
var

var.replace([1,2,3,4,5,6,7,8,9],value = 100)
var

var.replace("[A-Za-z]","python",regex=True)

var.replace({"SYMBOL":"[A-Z]"},22,regex=True)

var.replace(1,method="ffill")
var.replace(1,method="bfill")

var.replace(1,method="ffill",limit=2)

var.replace(1,method="ffill",limit=2,inplace=True)

```

```python
var = pd.read_csv("E:\Pandas\Video 10\demo.csv")
print(var)

var.interpolate(method="linear",axis=0)
var.interpolate(limit=2,limit_direction="forward")
var.interpolate(limit_area="backward")



```
